```python
#8 PUZZLE
# Fuction to accept user input of the board
import copy

def getInvCount(arr):
    inv_count = 0
    empty_value = -1 # to skip empty value
    for i in range(0, 9):
        for j in range(i + 1, 9):
            if arr[j] != empty_value and arr[i] != empty_value and arr[i] > arr[j]:
                inv_count += 1
    return inv_count
 #even number of inversions is solvable
 #odd number is odd
     
# This function returns true
# if given 8 puzzle is solvable.
def isSolvable(puzzle) :
 
    # Count inversions in given 8 puzzle
    inv_count = getInvCount([j for sub in puzzle for j in sub])
 
    # return true if inversion count is even.
    return (inv_count % 2 == 0)

def get_puzzle(message):
	print(message)
	board = []
	for i in range(3):
		row = input("Enter rows of the board (0-8), separated by spaces: ")
		row = [int(x) for x in row.split()]
		board.append(row)
	return board
 	
# To print the board
def print_board(board):
	for row in board:
		for item in row:
			print(item,end=" ")
		print()

#To get the position of 0 in the board
def get_empty(board):
	empty_pos = set()
	for i in range(3):
		for j in range(3):
			if board[i][j] == 0:
				empty_pos = (i,j)
	return empty_pos
	
	
# To get the possible moves of the 0
def get_moves(board):
	row, col = get_empty(board)
	moves = []
	if row > 0:
		moves.append((row-1, col))
	if col > 0:
		moves.append((row, col-1))
	if row < 2:
		moves.append((row+1, col))
	if col < 2:
		moves.append((row, col+1))
	return moves
 
# To calcualte the heuristics
def get_H(board,Goal):
	H = 0
	for i in range(3):
		for j in range(3):
			if board[i][j] != Goal[i][j]:
				H +=1
	return H

def move_empty(move,board):
	a,b = get_empty(board)
	x,y = move
	board[x][y],board[a][b] = board[a][b],board[x][y]
	return board
	
def solve(S,G):
	i=0
	visited=[]
	while S != G:	
		
		moves = get_moves(S)
		min_H = 9999
		best_move = None
		# print(S)
		for move in moves:
			temp=copy.deepcopy(S)
			temp = move_empty(move,temp)
			if temp not in visited:
				visited.append(temp)
				H = get_H(temp,G)
				if H < min_H:
					min_H = H
					best_move = move
					print(temp)
		S = move_empty(best_move,S)
		print("moved")
		print_board(S)
		i+=1
	print("solved")		
	print_board(S)
	return None
		
#if __name__ == '__main__':
start_board = get_puzzle("Enter the starting board")
#goal_board = get_puzzle("Enter the goal board")
	#start_board = [[1,2,3],[4,0, 6],[7,5,8]]
goal_board = [[1,2,3],[4,5,6],[7,8,0]]
if isSolvable(start_board):
	solve(start_board,goal_board)
else:
	print("Impossible to reach goal state")
	
```
